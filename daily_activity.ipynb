```python
import pandas as pd
import matplotlib.pyplot as plt

# Check if the file exists in the specified path
!ls Data/

# Assuming the file exists, read the CSV
df = pd.read_csv('time_tracker.csv')
print(df)

plt.figure(figsize=(8,8))
plt.pie(df['Duration (hours)'], labels=df['Activity'], autopct='%1.1f%%', startangle=90)
plt.title('Time Spent on Daily Activities')
plt.show()

plt.figure(figsize=(10,5))
plt.bar(df['Activity'], df['Duration (hours)'], color='pink')
plt.title('Hours Spent on Each Activity')
plt.xlabel('Activity')
plt.ylabel('Duration (hours)')
plt.xticks(rotation=45)
plt.show()
```
